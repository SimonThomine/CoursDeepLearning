# 使用 Diffusers 生成图像


在本笔记中，我们将学习如何使用 Hugging Face 的 **Diffusers** 库来操作扩散模型。由于这些模型计算资源需求较大，我们将通过一个轻量级模型的小示例进行演示。


### 实现步骤


我们使用 OFA-Sys 的 **[small-stable-diffusion](https://arxiv.org/pdf/2206.14366)** 模型（模型标识：`OFA-Sys/small-stable-diffusion-v0`）。


```python
from diffusers import DiffusionPipeline

diffuser = DiffusionPipeline.from_pretrained("OFA-Sys/small-stable-diffusion-v0") #"stabilityai/sdxl-turbo"
```

生成过程可能需要数分钟，具体时间取决于您的计算机性能。


```python
outputAutre=diffuser("A car in the winter")
```

```python
import matplotlib.pyplot as plt
print(outputAutre.images[0])
plt.imshow(outputAutre.images[0])
plt.axis("off")
plt.show()
```

现在，您已经学会如何使用 Hugging Face 的 **Diffusers** 库生成图像。
